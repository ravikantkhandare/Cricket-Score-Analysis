```python
data["Player_Name"]=data["Player"].str.split("(").str[0].str.strip()
```

```python
data["Country"]=data["Player"].str.split("(").str[1].str.split(")").str[0].str.strip()
data.head()
```

```python
data.drop("Player",axis=1,inplace=True)
```

```python
data["Start_Year"]=data["Span"].str.split("-").str[0]
```

```python
data["End_Year"]=data["Span"].str.split("-").str[1]
```

```python
data.drop("Span",axis=1,inplace=True)
```

```python
data.head(1)
```

```python
data.info()
```

```python
data[["Start_Year","End_Year"]]=data[["Start_Year","End_Year"]].astype("int")
```

```python
data["Carrier_Lenth"]=data["End_Year"]-data["Start_Year"]
```

```python
data.head(2)
```

```python
data["Mat"]=data["Mat"].replace("[*]","",regex=True)
data["Mat"]=data["Mat"].astype("int")
```

```python
data["HS"]=data["HS"].replace("[*]","",regex=True)
data["HS"]=data["HS"].astype("int")
```

```python
data["BF"]=data["BF"].replace("[+]","",regex=True)
data["BF"]=data["BF"].replace("[-]",0,regex=True)
data["BF"]=data["BF"].astype("int")
```

```python
data["SR"]=data["SR"].replace("[-]",0,regex=True)
data["SR"]=data["SR"].astype("float")
```
